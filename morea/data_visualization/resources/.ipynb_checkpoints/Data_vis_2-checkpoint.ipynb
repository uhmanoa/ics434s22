```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
## Common Plot Types
```

### line plot

* Data points plotted as a line
  * line can have different styles
  * Solid by default, and add different styles when plotting more than one line
  
* Typically used when the x-axis is contibuous or when it's acceptable to make such an assumption
  * ex.: `x` time series data

```python
plt.figure(figsize=(12,6))

year = np.arange(0,100)
profit = np.random.randint(-2,80, 100)
plt.plot(year, profit)
```

### Scatter plot

* Data points plotted as dots, or circles or other symbols across two axes. 
  * Symbols can use different shape to represent a third, typically categorical variable
  * Symbold can use size of hue to represent a thurd continue variable
* typically used when the x-axis is not continuous or when such assumption does not make sense.
  * ex.: `x` A few categorical obsevations, sparse time data, etc..
    

```python
plt.figure(figsize=(12,6))
age = np.random.randint(20, 45, 100)
income = np.random.randint(45, 165, 100)

plt.scatter(age, income)
```

```python
plt.figure(figsize=(12,6))
age = np.random.randint(20, 45, 100)
income_male = np.random.randint(45, 165, 100)
income_female = np.random.randint(45, 115, 100)
plt.scatter(age, income_male, label= "male income")
plt.scatter(age, income_female, label= "female income")
plt.legend()
```

```python
### Area Chart


The idea of an area chart is based on the line chart. Colored regions (areas) show us the development of each variable over time.
There are three types of area charts: regular area chart, stacked area chart, and 100% stacked area chart.
When to use an area chart
We use area charts to show how the parts of a complete change over time. For example, the happiness score has six generating divisions; we would like to see each of these divisions’ contributions.
Moreover, if we are interested in the portion generated by each division and not that much of the total amount of the division self, we can use a 100% stacked area chart. This will show each division’s percentage contribution over time.
When to avoid area charts
Area charts are not the best choice if we want to compare the size of different shares with each other. If you want to show that one share overtook another one; or if the differences between our values are very small, consider a line chart instead.
```

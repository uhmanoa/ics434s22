````python
import numpy as np

import matplotlib.pyplot as plt
from scipy.stats import poisson

````

### The Poisson Distribution

* Used in experiments that model the numbers of events in a fixed period of time or a fixed area of space
* Ex.: 
    - Number of daily accidents on the H1
    - Number of people who enter a grocery store every hour
    - Number of childbirths in Hawaii every day
    - Number of chocolate chips in a cookie
    - etc.


#### Properties of the Poisson Distribution

* Poisson has only one parameter that determines its shape and location.
  * The parameter is referrd to as $\lambda$

* A random variable $X$ is Poisson, written $X \sim Poisson(\lambda)$, if:

  * $X$ is the number of events in a given time period or space area
    * Countably infinite positive values [0, 1, ... ]
    
  * The events are independent (The occurrent of events does not increase the likelihood of other events occurring)

  * The mean and the variance of a Poisson are represented using the same parameter $\lambda$.
  
* The $pmf$ of a Poisson is:

$$
p(X=x) = \frac{e^{_\lambda} \lambda^x}{x!}
$$

#### Example

- If the clientele visiting a store per hour follows a Poisson distribution with $\lambda = 10$. 

- The probability distribution of the random variable $X$ representing the number of clients in the store is:

````python
plt.figure(figsize=(12,6))
x = np.arange(0, 30, 1)
pmfs_x = poisson.pmf(x, 10)
plt.plot(x, pmfs_x,)
_ = plt.xlabel("$X$")
_ = plt.ylabel("$P(X)$")
````

### Link between the mean and the variance

- As mentioned, the mean and variance are both modeled with the same parameter $\lambda$
  - What is the significance of this statement?

````python

plt.figure(figsize=(16,6))

x = np.arange(0, 23, 1)
pmfs_x = poisson.pmf(x, 10)
plt.plot(x, pmfs_x)

x = np.arange(160, 240, 1)
pmfs_x = poisson.pmf(x, 200)
plt.plot(x, pmfs_x)


x = np.arange(350, 450, 1)
pmfs_x = poisson.pmf(x, 400)
plt.plot(x, pmfs_x)
````

### Notes about the Poisson
* For values of $\lambda > 40$, the Poisson can be approximated as Gaussian with $\mu=$ and $\sigma = \sqrt{\lambda}$
  * This can be shown mathematically and this simplification sometimes makes the math easy 
  
* Not all count data is Poisson
  * Data where the mean is not equal to the variance is not Poisson
  * Occurs in overdispersed data when external factors, such as noise, contribute to the observed variance.
    * Negative Binomial is often used with such data
      * More on this distribution later.

### Computing the Probability of Observing these Data 

* We can compute the probability given each possible parameter  
  * We limit ourselves to reasonable values
  

  * We assume the data is generated from a Poisson distribution
  * We know how to compute the probability (pmf) of a single event
          `poisson.pmf(x, lambda)`
          
  * Since all events are independent, the probability of the complete dataset is simply the product of all probabilities
              `np.product(poisson.pmf(x, lambda))`
  



### Compound probability of Independent Events

* Given a fair coin $p(H)=p(T) = 1/2$
* If we flip the coin twice, the sample sapce is \{HH, HT, TH, TT\} and  probabilities of the different outcomes is:

$$
p(HH) = p(H) * p(H) = 1/4 \\
p(HT) = p(H) * p(T) = 1/4 \\
p(TH) = p(T) * p(H) = 1/4 \\
p(TT) = p(T) * p(T) = 1/4 \\
$$

* As expected, all the probabilities sum to one 

* This only works if the probabilities are independent; i.e., obtaining `head` in the first toss does not affect the outcome of the second coin flip

### Computing the Likelihood (Probability)

* If the events are independent, the probability of the data is is the product of the probabilities

* This probability is the likelihood function $L$ of $\lambda$

$$ 
\begin{align}
L([x_1,x_2,...x_{90}] | \lambda) &= pmf(x_1, \lambda) \times  pmf(x_2, \lambda)  ....  pmf(x_{90}, \lambda) \\
  &=  \prod_{i=1}^{90} pmf(x_i, \lambda) 
\end{align}
$$

### Understanding the Likelihood - Cont'd


* Working with small numbers can lead to likelihood to be very small
  * Can lead to arithmetic floating point underflow 
    * Results of a calculation smaller than what can actually be represented in memory
* Rather than computing likelihood, we can calculate Log-likelihood
  * Scale changes but share of the function does not
  * Those are no longer probabilities as they don't sum to 1

We know that:

$$
log(x \dot y) = log(x) + log(y)
$$

Therefore:


\begin{align}
log(L([x_1,x_2,...x_90] | \lambda)) &= log(pmf(x_1, \lambda)) \times  log(pmf(x_2, \lambda))  ....  log(pmf(x_{90}, \lambda)) \\
  &=  \sum_{i=1}^{90} log(pmf(x_i, \lambda)) 
\end{align}


````python
plt.figure(figsize=(12,8))

x = range(12, 21)
l_x = [np.sum(np.log(poisson.pmf(citations_data["Counts"], i))) for i in x]

plt.scatter(x, l_x)
````

###  Maximum Likelihood Estimation

$$

* Recall that ML estimators are simply point-estimates that maximize the likelihood of data

* Do not give us a probability of the value we generate
 * We do not have the probability or confidence that it is the correct parameter of the population


### Parameter Estimation and Variance


* If we have the relevant data for the entire population, we can simply calculate the parameter. 
* We know that a statistic  based on a random sample can be a reasonable estimate of an unknown parameter in the population
 * BUT: We cannot guarantee that two samples drawn at random will have exctly the same paramter
 
 
* The questions i then: ** "How different could this estimate have been, if the sample had come out differently?"**
 

###  Sampling and Resampling

* A random sample is just one of numerous possible random samples

  * Our parameter estimate is, therefore, just one of numerous plausible estimates


* By how much could those estimates vary? 

* Good was to test this is by drawing another sample from the population and compare the new estimate to the old one

  * Sampling from the population can be costly, tedious or simply not possible



### The Bootstrap: Resampling from the Sample ###

* Key Idea: A large random sample is likely to resemble the population from which it is drawn. 

 * This observation allows data scientists replicate the sampling procedure by *sampling from the sample*. 

* Here are the steps of *the bootstrap method* for generating another random sample that resembles the population:

- **Treat the original sample as if it were the population.**

- **Draw from the sample**, at random **with** replacement, **the same number of times as the original sample size**

* For instance, rather than sample new citation, we consider our dataset as a population  and draw new samples from it

````python
test_data = [1,2,3,4]

np.random.choice(test_data, 4, replace=True)

# np.random.choice uses replace = True by default
````

````python
plt.figure(figsize=(12,8))



plt.hist(citations_data["Counts"], bins=8, normed=True, edgecolor='black', linewidth=1.2, alpha =0.5)
x = np.arange(30)
bootstrap_1 = np.random.choice(citations_data["Counts"], len(citations_data["Counts"]))
_ = plt.hist(bootstrap_1, bins=8, normed=True, edgecolor='black', linewidth=1.2, alpha =0.5)

````

### Remember 
1 - Resample the same number of times as the original sample size.  

  - The reason is that the variability of an estimate depends on the size of the sample. 
  - To see how different the sample could have been, we have to compare it to the median of other samples of size 500.

- If we drew 500 times at random *without* replacement from our sample of size 500, we would just get the same sample back. 

  - By drawing *with* replacement, we create the possibility for the new samples to be different from the original, because some employees might be drawn more than once and others not at all.


### Why is This a Good Idea? 

* Similarity by transitivity

  *  By the law of averages, the distribution of the original sample is likely to resemble the population, 

  *  The distributions of all the "resamples" are likely to resemble the original sample.

  * So the distributions of all the resamples are likely to resemble the population as well.
